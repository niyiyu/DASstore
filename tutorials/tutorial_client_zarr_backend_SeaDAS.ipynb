## Client class
A `Client` class can be instantiated to access a specific DASstore dataset. Users would specify the `bucket_name` and the `endpoint_url` of the object storage. Here, we store one month of SeaDAS data in the `seadas-december-2022` bucket in an object storage running at port `9000` of a local server. 

* `bucket   = "seadas-december-2022"`
* `endpoint = "dasway.ess.washington.edu:9000"`
* `anon = False` (by default)
* `secure = False` (by default)

````python
client = Client("seadas-december-2022", 
                "dasway.ess.washington.edu:9000", 
                anon = False, secure = False)
````

## Access Control

As shown above, MinIO allows access control, and only when credentials are provided will allow access to the bucket when the data are not public. Users would need to have an access `key/secure` pair for authentication (a credential). The key pair is stored in `~/.dasstore/credential` by default, but this is subject to change in the future.
```
$ cat ~/.dasstore/credentials
[ENDPOINT_URL]
aws_access_key_id = ACCESS_KEY_ID
aws_secret_access_key = ACCESS_KEY_SECRET
```
Users usually get into the error at the first use, which indicates that the credential is not found at a specific place (`~/.dasstore/credential` by default). We can manually add the credential for a specfic `endpoint` using the function from `dasstore.utils.credential.add_credential`, e.g.:
* `endpoint = "dasway.ess.washington.edu:9000"`

**Note that the ACCESS_KEY_ID/ACCESS_KEY_SECRET shown below should be replaced by the actual key pair received from the MinIO administrator.**

````python
from dasstore.utils.credential import add_credential
add_credential("dasway.ess.washington.edu:9000")
````

With the proper credential, users are able to connect to the bucket.

````python
client = Client("seadas-december-2022", 
                "dasway.ess.washington.edu:9000", 
                anon = False)
````

## client.get_data(channel_index, starttime, endtime)
The `get_data` function of the client can be used to get the raw DAS data. The channel index can be either an integer (single channel) or an array (multiple channels). The `starttime` and `endtime` are just the string UTC time.

When converting DAS data into the DASStore, users can specify the fill value (Not-a-Number, NaN in this case). This feature is useful in showing data gaps (or any potential issue that affected the writing).

**Note:** as Zarr chunks DAS data (1 minute, 50 channels in this case), it's recommended to request multiple channels at the same time, instead of sequentially request single channel. Remember that for Zarr, always **the whole data chunk** is being requested and received, although you may not want them.


````python
# get one hour data of channel index 1
# note that index starts from 1 (Python convension)
t0 = time.time()
data = client.get_data(1, 
                       starttime = "2022-12-31T10:00:00.000", 
                       endtime = "2022-12-31T11:00:00.000")

print("Query time:\t%.3f second" % (time.time() - t0))
print(f"Data shape:\t {data.shape}")

plt.figure(figsize = (12, 4), dpi = 100)
plt.plot(data)

# note the data gap around t = 300000 (50 minutes after the starttime)
````

````python
# get half-hour data of channel index 500 to 1100

t0 = time.time()
data = client.get_data(channels = np.arange(500, 1100), 
                       starttime = "2022-12-15T13:30:00.000", 
                       endtime   = "2022-12-15T14:00:00.000")
print("Query time: %.3f second" % (time.time() - t0))
print(f"Data shape:\t {data.shape}")

plt.figure(figsize = (12, 4), dpi = 300)
plt.imshow(data, aspect = 'auto', cmap = 'RdBu', vmax = 2, vmin = -2)
plt.xlabel("Time (min)", fontsize = 15)
plt.ylabel("Channel index", fontsize = 15)
_ = plt.yticks(np.linspace(0, 600, 7), [int(i) + 500 for i in np.linspace(0, 600, 7)])
_ = plt.xticks(np.linspace(0, 180000, 7), [int(i) for i in np.linspace(0, 180000, 7)/6000])
    
# note the data gap for all channels.
````

## One month data: 50 chunked channels

This test shows querying one month of data from SeaDAS experiment. 
For the 50-channel experiment, there are 52.1 GB data being queried in theory.

````python
# get one day of data for channel index 0 - 50 (chunked)
# note that index starts from 1 (Python convension)
t0 = time.time()
data = client.get_data(np.arange(50), 
                       starttime = "2022-12-01T00:00:00.000", 
                       endtime = "2023-01-01T00:00:00.000")

print("Query time:\t%.3f second" % (time.time() - t0))
print(f"Data shape:\t {data.shape}")
````
